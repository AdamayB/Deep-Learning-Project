Standardize the data for T-SNE and PCA

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df.select_dtypes(include=[np.number]))
```

T-SNE

```python
tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(scaled_data)
```

PCA

```python
pca = PCA(n_components=2)
pca_results = pca.fit_transform(scaled_data)
```

```python
pca_full = PCA()
pca_full.fit(scaled_data)
feature_importance = np.abs(pca_full.components_[0])
important_features = df.columns[np.argsort(feature_importance)[::-1]]
print("Most Important Features:\n", important_features)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance[np.argsort(feature_importance)[::-1]], y=important_features)
plt.title('Feature Importance from PCA')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()
```

Creating plots

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is already loaded, replace 'df' with your actual DataFrame variable if different
# features to plot
features = ['Age', 'Overeating or loss of appetite', 'Problems concentrating or making decision',
            'Feeling anxious', 'Irritable towards baby & partner', 'Trouble sleeping at night',
            'Suicide attempt', 'Problems of bonding with baby', 'Feeling sad or Tearful',
            'Feeling of guilt']


print("Columns in DataFrame:", df.columns)


target_column = 'Feeling anxious'

num_features = len(features)
num_rows = (num_features + 2) // 3
fig, axs = plt.subplots(num_rows, 3, figsize=(18, 6 * num_rows))
axs = axs.flatten()

for i, feature in enumerate(features):
    if feature not in df.columns:
        print(f"Feature '{feature}' is not present in the DataFrame")
        continue
    try:
        df_plot = df.groupby([feature, target_column]).size().unstack().fillna(0)
        df_plot.plot(kind='bar', ax=axs[i])
        axs[i].set_title(feature.replace("_", " ").capitalize())
        axs[i].set_xlabel(feature.replace("_", " ").capitalize())
        axs[i].set_ylabel('Count')
        axs[i].set_xticklabels(df_plot.index, rotation=0)
    except KeyError as e:
        print(f"Error processing feature '{feature}': {e}")
    except Exception as e:
        print(f"An unexpected error occurred while processing feature '{feature}': {e}")

for j in range(i + 1, len(axs)):
    fig.delaxes(axs[j])

plt.tight_layout()
plt.show()
```
