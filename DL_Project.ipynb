```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.columns
```

```python
x=data[['Age', 'Feeling sad or Tearful', 'Irritable towards baby & partner',
       'Trouble sleeping at night',
       'Problems concentrating or making decision',
       'Overeating or loss of appetite', 'Feeling of guilt',
       'Problems of bonding with baby', 'Suicide attempt']]
y=data['Feeling anxious']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Bidirectional, Dense
from tensorflow.keras.regularizers import l2

import tensorflow as tf
from tensorflow.keras.metrics import Recall

model = tf.keras.Sequential()
model.add(tf.keras.layers.Reshape((9, 1), input_shape=(9,)))
model.add(Bidirectional(tf.keras.layers.LSTM(128)))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))


model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy', Recall()])



```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python

history_dense = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test),callbacks=[early_stopping])
```

```python
plt.plot(history_dense.history['loss'], label='Training Loss')
plt.plot(history_dense.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.plot(history_dense.history['accuracy'], label='Training Accuracy')
plt.plot(history_dense.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(X_test)


y_pred_binary = (y_pred >= 0.4).astype(int)
```

```python
import numpy as np
import tensorflow as tf
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


model = tf.keras.models.load_model('/content/DL_PROJECT_II.h5')


y_pred_probs = model.predict(X_test)


threshold = 0.4
y_pred = (y_pred_probs >= threshold).astype(int)


accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Accuracy: {accuracy * 100:.2f}%')
print(f'Precision: {precision* 100:.4f}%')
print(f'Recall: {recall* 100:.4f}%')
print(f'F1-score: {f1* 100:.4f}%')
```
