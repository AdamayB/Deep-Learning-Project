```python
from sklearn.model_selection import train_test_split
```

```python
x=data[['Age', 'Feeling sad or Tearful', 'Irritable towards baby & partner',
       'Trouble sleeping at night',
       'Problems concentrating or making decision',
       'Overeating or loss of appetite', 'Feeling of guilt',
       'Problems of bonding with baby', 'Suicide attempt']]
y=data['Feeling anxious']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Bidirectional, Dense
from tensorflow.keras.regularizers import l2
# Define the model
# model = tf.keras.Sequential()
# model.add(tf.keras.layers.Dense(10, activation='relu'))
# model.add(tf.keras.layers.Dense(64, activation='relu'))
# model.add(tf.keras.layers.Dense(32, activation='relu'))
# model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
# Define the model
import tensorflow as tf
from tensorflow.keras.metrics import Recall

# Define the model
model = tf.keras.Sequential()

# Reshape the input to (batch_size, timesteps=9, features=1)
model.add(tf.keras.layers.Reshape((9, 1), input_shape=(9,)))

# LSTM layer to process the input
model.add(Bidirectional(tf.keras.layers.LSTM(128)))

# Output layer with sigmoid activation for binary classification
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy', Recall()])



```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
# Assuming X_train, y_train, X_val, y_val are your training and validation data
history_dense = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test),callbacks=[early_stopping])
```

```python
plt.plot(history_dense.history['loss'], label='Training Loss')
plt.plot(history_dense.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot training and validation accuracy
plt.plot(history_dense.history['accuracy'], label='Training Accuracy')
plt.plot(history_dense.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(X_test)  # This will give you probabilities

# Apply threshold to convert probabilities to 0 or 1
y_pred_binary = (y_pred >= 0.4).astype(int)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy_fusion = accuracy_score(y_test, y_pred_binary)
precision_fusion = precision_score(y_test, y_pred_binary, average='weighted')
recall_fusion = recall_score(y_test, y_pred_binary, average='weighted')
f1_fusion = f1_score(y_test, y_pred_binary, average='weighted')

# Print the results
print(f"Accuracy: {accuracy_fusion:.6f}")
print(f"Precision: {precision_fusion:.6f}")
print(f"Recall: {recall_fusion:.6f}")
print(f"F1 Score: {f1_fusion:.6f}")
```
